Para el objetivo del proyecto, hay que trabajar solo con 6 países: Argentina, Chile, Colombia, Mexico, Perú, Brasil. 

```python
paises_proyecto= ['Argentina','Chile','Colombia','Mexico','Peru','Brazil']
data_latinoamerica=df_latinoamerica[df_latinoamerica['country_name'].isin(paises_proyecto)]
# isin filtra la columna country name a partir de la lista con los paises seleccionados
```

```python
#Verifico si el numero de rows disminuye por el filtrado realizado
np.shape(data_latinoamerica)
```

```python
#Verifico la cantidad de nulos o valores faltantes en las columnas
data_latinoamerica.isnull().sum()[data_latinoamerica.isnull().sum()>4000000]
#Entre corchetes hago una mascara que va a filtrar la suma de valores nulos por colummna, y solo va a mostrar cuando
#los nulos son superiores a los 4000000,lo que es un valor preocupante de datos faltantes
```

```python
#Realizo un nuevo filtrado a partir de location_key, utilizando solo las instancias de AR,CL,CO,MX,PE,BR
data_paises_latinoamerica=data_latinoamerica[data_latinoamerica['location_key'].isin(['AR','CO','CL','MX','PE','BR'])]
```

```python
#Verifico si disminuyeron el número de rows
np.shape(data_paises_latinoamerica)
```

```python
#Filtro la data de los países seleccionados a partir de la fecha 2021-01-01
data_latinoamerica_fecha= data_paises_latinoamerica[data_paises_latinoamerica['date']>'2021-01-01']
```

```python
#Verifico la disminucion del número de rows
np.shape(data_latinoamerica_fecha)
```

```python
#Conteo de valores nulos en la variable de las fechas seleccionadas para su posterior eliminación corrección
valores_nulos_fecha=data_latinoamerica_fecha.isnull().sum()[data_latinoamerica_fecha.isnull().sum()>0]
print(valores_nulos_fecha)
```

Usar bucles for para completar con datos de la media los datos nulos de las columnas para cada país

```python
for pais in  paises_proyecto:
 data_latinoamerica_fecha['new_confirmed']=data_latinoamerica_fecha['new_confirmed'].fillna(data_latinoamerica_fecha[data_latinoamerica_fecha['country_name']==pais]['new_confirmed'].mean())
 data_latinoamerica_fecha['new_deceased']=data_latinoamerica_fecha['new_deceased'].fillna(data_latinoamerica_fecha[data_latinoamerica_fecha['country_name']==pais]['new_deceased'].mean())
 data_latinoamerica_fecha['cumulative_confirmed']=data_latinoamerica_fecha['cumulative_confirmed'].fillna(data_latinoamerica_fecha[data_latinoamerica_fecha['country_name']==pais]['cumulative_confirmed'].mean())
 data_latinoamerica_fecha['cumulative_deceased']=data_latinoamerica_fecha['cumulative_deceased'].fillna(data_latinoamerica_fecha[data_latinoamerica_fecha['country_name']==pais]['cumulative_deceased'].mean())
 data_latinoamerica_fecha['cumulative_vaccine_doses_administered']=data_latinoamerica_fecha['cumulative_vaccine_doses_administered'].fillna(0)
 data_latinoamerica_fecha['average_temperature_celsius']=data_latinoamerica_fecha['average_temperature_celsius'].fillna(data_latinoamerica_fecha[data_latinoamerica_fecha['country_name']==pais]['average_temperature_celsius'].mean())
 data_latinoamerica_fecha['minimum_temperature_celsius']=data_latinoamerica_fecha['minimum_temperature_celsius'].fillna(data_latinoamerica_fecha[data_latinoamerica_fecha['country_name']==pais]['minimum_temperature_celsius'].mean())
 data_latinoamerica_fecha['maximum_temperature_celsius']=data_latinoamerica_fecha['maximum_temperature_celsius'].fillna(data_latinoamerica_fecha[data_latinoamerica_fecha['country_name']==pais]['maximum_temperature_celsius'].mean())
 data_latinoamerica_fecha['rainfall_mm']=data_latinoamerica_fecha['rainfall_mm'].fillna(data_latinoamerica_fecha[data_latinoamerica_fecha['country_name']==pais]['rainfall_mm'].mean())
 data_latinoamerica_fecha['relative_humidity']=data_latinoamerica_fecha['relative_humidity'].fillna(data_latinoamerica_fecha[data_latinoamerica_fecha['country_name']==pais]['relative_humidity'].mean())
 data_latinoamerica_fecha['new_recovered']=data_latinoamerica_fecha['new_recovered'].fillna(0)
 data_latinoamerica_fecha['cumulative_recovered']=data_latinoamerica_fecha['cumulative_recovered'].fillna(0)

#En el caso de las dosis suministradas, y el acumulado de estas dosis, si no se poseen datos, se rellena con 0 porque se considera que no se han suministrado dosis
```

```python
#Finalmente verificamos nuevamente las registros vacios y el tamaño del nuevo data set
print(np.shape(data_latinoamerica_fecha))
print(data_latinoamerica_fecha.isnull().sum()[data_latinoamerica_fecha.isnull().sum()>0])
```

```python
#Guardar los datos filtrados en un archivo csv para en su posterior uso, no tener que volver a realizar la limpieza
data_latinoamerica_fecha.to_csv('DatosFinalesFiltrados.csv',index=False)
```
