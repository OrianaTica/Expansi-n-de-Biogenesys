```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
archivo_integrador=('DatosFinalesFiltrados.csv')
df_latinoamerica_filtrada=pd.read_csv(archivo_integrador)
df_latinoamerica_filtrada.head(7)
```

```python
df_latinoamerica_filtrada.info()
```

```python
#Se formatea la columna de 'date' que era del tipo object al tipo de formato de fecha
#Se modifica el Ã­ndice del data set, estableciendo la columna 'date' como nuevo Ã­ndice
df_latinoamerica_filtrada['date']=pd.to_datetime(df_latinoamerica_filtrada['date'])
df_latinoamerica_filtrada=df_latinoamerica_filtrada.set_index('date')
df_latinoamerica_filtrada.head(7)
```

```python
#A partir del indice de fechas, se extrae la semana y se crea una nueva columna que posee la informacion de la semana de 'date'
df_latinoamerica_filtrada['semana']=df_latinoamerica_filtrada.index.isocalendar().week
df_latinoamerica_filtrada.head()
```

Con los datos limpios del dataset se usa una mÃ¡scara para utilizar solo los datos del 2021

```python
df_latinoamerica_paises=df_latinoamerica_filtrada.loc[
    (df_latinoamerica_filtrada.index>= '2021-01-01') & 
    (df_latinoamerica_filtrada.index<= '2021-12-31')
]
df_latinoamerica_paises
```

Con los registros diarios, se busca conocer la suma de casos diarios confirmados y  de muertes semanales diferenciando por paÃ­s seleccionado.

```python
casos_semanales=df_latinoamerica_paises.groupby(['country_name','semana'])[['new_confirmed','new_deceased']].sum().reset_index()
casos_semanales
```

```python
#Graficar la evolucion de los casos semanal por pais
fig, ax = plt.subplots(2, 1, figsize=(10, 10))
for country in casos_semanales['country_name'].unique(): 
    casos_semanales_pais=casos_semanales[casos_semanales['country_name']==country]
    ax[0].plot(casos_semanales_pais['semana'],casos_semanales_pais['new_confirmed'],label=f'{country} New confirmed')
    ax[1].plot(casos_semanales_pais['semana'],casos_semanales_pais['new_deceased'],label=f'{country} New deceased')
    
ax[0].set_xlabel('Semana del aÃ±o')
ax[0].set_ylabel('Casos nuevos COVID-19')
ax[0].set_title('EvoluciÃ³n semanal de casos nuevos de COVID 19')
ax[0].legend(bbox_to_anchor=(1.05,1),loc='upper left')

ax[1].set_xlabel('Semana del aÃ±o')
ax[1].set_ylabel('Muertes por COVID-19')
ax[1].set_title('EvoluciÃ³n semanal de muertes')
ax[1].legend(bbox_to_anchor=(1.05,1),loc='upper left')

plt.tight_layout()

    
    
```

```python
#Creacion de una nueva columna de 'mes' con informaciÃ³n extraÃ­da del Ã­ndice 'date'
df_latinoamerica_paises['mes']=df_latinoamerica_paises.index.month
#Calculo de los casos mensuales por pais
casos_mensuales=df_latinoamerica_paises.groupby(['country_name','mes',])[['new_confirmed','new_deceased']].sum().reset_index()
fig, ax = plt.subplots(2, 1, figsize=(10, 10))
for country in casos_mensuales['country_name'].unique(): 
    casos_mensuales_pais=casos_mensuales[casos_mensuales['country_name']==country]
    ax[0].plot(casos_mensuales_pais['mes'],casos_mensuales_pais['new_confirmed'],label=f'{country} New confirmed')
    ax[1].plot(casos_mensuales_pais['mes'],casos_mensuales_pais['new_deceased'],label=f'{country} New deceased')
    
ax[0].set_xlabel('Mes del aÃ±o')
ax[0].set_ylabel('Casos nuevos COVID-19')
ax[0].set_title('EvoluciÃ³n mensual de casos nuevos')
ax[0].legend(bbox_to_anchor=(1.05,1),loc='upper left')

ax[1].set_xlabel('Mes del aÃ±o')
ax[1].set_ylabel('Muertes por COVID-19')
ax[1].set_title('EvoluciÃ³n mensual de muertes')
ax[1].legend(bbox_to_anchor=(1.05,1),loc='upper left')
```

A partir de los grÃ¡ficos de la evoluciÃ³n de los casos en el tiempo,se concluye que el paÃ­s con mayor cantidad de infectados es Brasil,seguido por Argentina y Colombia. Analizando la cantidad de muertes Brasil tambiÃ©n es el paÃ­s con mayor cantidad de muertes, lo que para poder entender su origen se debe conocer la distribuciÃ³n etaria de los contagiados, como fue el manejo de la pandemia y condiciones o patologÃ­as preexistentes.

```python
#Comparacion de progreso de vacunacion entre paises
data_vacunacion=df_latinoamerica_filtrada.groupby(['country_name'])[['cumulative_vaccine_doses_administered']].max().sort_values(by='cumulative_vaccine_doses_administered', ascending=False)
plt.figure(figsize=(12,6))
sns.barplot(x= data_vacunacion.index , y=data_vacunacion['cumulative_vaccine_doses_administered'], hue=data_vacunacion.index, palette='viridis')
plt.title('Progreso de la vacunacion por pais')
plt.xlabel('PaÃ­s')
plt.ylabel('DÃ³sis de vacunas')
```

Al analizar el progreso de la vacunaciÃ³n por paÃ­s, se observa que Brasil es el territorio con mÃ¡s dÃ³sis de vacunas suministradas, lo que podrÃ­a deberse a la gran cantidad de contagiados. El segundo paÃ­s con mÃ¡s dÃ³sis es Mexico, seguido por Argentina.

Buscando conocer como afecta la vacunaciÃ³n en la recuperaciÃ³n y la cantidad de casos activos, se procede a una comparaciÃ³n de casos activos vs recuperados por paÃ­s. 

```python
#Visualizacion de los casos activos vs los casos recuperados
paises_proyecto= ['Argentina','Chile','Colombia','Mexico','Peru','Brazil']
df_latinoamerica_filtrada['active_cases']=(df_latinoamerica_filtrada['cumulative_confirmed']-df_latinoamerica_filtrada['cumulative_deceased']-df_latinoamerica_filtrada['cumulative_recovered'])
fig,ax= plt.subplots(figsize=(10,5))


sns.lineplot(data=df_latinoamerica_filtrada.resample('ME').mean(numeric_only=True), x='date', y='active_cases', label='Casos Activos')
sns.lineplot(data=df_latinoamerica_filtrada.resample('ME').mean(numeric_only=True), x='date', y='cumulative_recovered', label='Casos Recuperados')
plt.xlabel('Fecha')
plt.ylabel('Conteo')
plt.title('Casos activos vs Casos recuperados por paÃ­s')
plt.legend(bbox_to_anchor=(1.05,1),loc='upper left')
```

Hay un aumento notable de los casos activos a principios de 2022, lo que podrÃ­a indicar un brote o una nueva ola de infecciones. Posteriormente, los casos activos se estabilizan y muestran un ligero aumento.
La lÃ­nea de casos recuperados muestra un aumento constante durante todo el perÃ­odo, lo cual es tÃ­pico de la recuperaciÃ³n de una infecciÃ³n. Sin embargo, hay una pequeÃ±a desaceleraciÃ³n en el crecimiento alrededor de mediados de 2022.
Conociendo como se comporta la enfermedad se busca conocer como impactan las estrategias de vacunaciÃ³n.

```python
#Grafico que muestra la relacion entre la cobertura de vacunacion y la reduccion de los casos de COVID
df_latinoamerica_filtrada['porcentaje_vacunacion']=(df_latinoamerica_filtrada['cumulative_vaccine_doses_administered']/df_latinoamerica_filtrada['population'])*100

plt.figure(figsize=(10,5))
ax1=plt.gca()
ax2=ax1.twinx()

df_latinoamerica_filtrada.resample('ME').mean(numeric_only=True)['porcentaje_vacunacion'].plot(ax=ax1,color='purple',label='Porcentaje de vacunaciÃ³n')
df_latinoamerica_filtrada.resample('ME').mean(numeric_only=True)['new_confirmed'].plot(ax=ax2,color='red',label='Casos COVID')

ax.set_xlabel('Fecha')
ax1.set_ylabel('Porcentaje de vacunaciÃ³n')
ax2.set_ylabel('Casos de COVID')
ax1.set_title('Casos COVID vs Cobertura de vacunacion')
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')
```

Con el grÃ¡fico se concluye, que hay una correlaciÃ³n negativa entre el porcentaje de vacunaciÃ³n y los casos de COVID, es decir que cuando aumenta la vacunaciÃ³n los casos tienden a disminuir. La presencia de picos en el grÃ¡fico de los casos COVID puede indicar la presencia de olas de infecciÃ³n o cambios en las polÃ­ticas de salud pÃºblica que impacten en esta variable. 

Para conocer si el pico de contagiados se debe a un factores ambientales, se realiza un anÃ¡lisis de temperaturas medias vs los nuevos casos de COVID.

```python
#Grafico de la relacion entre la temperatura promedio y los nuevos casos confirmados de covid 
plt.figure(figsize=(20,7))
ax1=plt.gca()
ax2=ax1.twinx()
ax1.plot(df_latinoamerica_filtrada.resample('ME').mean(numeric_only=True).index,df_latinoamerica_filtrada.resample('ME').mean(numeric_only=True)['new_confirmed'],color='red',label='New confirmed')
ax2.plot(df_latinoamerica_filtrada.resample('ME').mean(numeric_only=True).index,df_latinoamerica_filtrada.resample('ME').mean(numeric_only=True)['average_temperature_celsius'],color='blue',label='Average temperature',linestyle='--')

ax1.set_xlabel('Fecha')
ax1.set_ylabel('Nuevos casos confirmados')
ax2.set_ylabel('Temperatura promedio (Celsius)')
ax1.set_title('Nuevos casos de COVID y temperatura promedio en el tiempo')
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.tight_layout()


```

Los casos confirmados y la temperatura promedio parecen tener patrones estacionales, en donde la temperatura promedio muestra una tendencia cÃ­clica con picos y valles que coinciden con los cambios estacionales. Se puede observar que con la disminuciÃ³n de la temperatura hay un leve aumento de casos confirmados.

Conociendo la cantidad de poblaciÃ³n contagiada, se busca conocer como la urbanizaciÃ³n afecta los casos de COVID, ya que se puede concluir que la mayor densidad poblacional en menos territorio es favorable para la propagaciÃ³n del virus.

```python
pd.set_option('display.float_format', '{:.2f}'.format)
#Impacto de la urbanizacion en la propagacion del covid
df_latinoamerica_filtrada['porcentaje_poblacion_urbana']=df_latinoamerica_filtrada['population_urban']/df_latinoamerica_filtrada['population']
df_latinoamerica_filtrada['porcentaje_poblacion_confirmada']=df_latinoamerica_filtrada['cumulative_confirmed']/df_latinoamerica_filtrada['population']
casos_urbanos = df_latinoamerica_filtrada.groupby('country_name').last().sort_values(by='porcentaje_poblacion_urbana')
casos_urbanos

plt.figure(figsize=(10,6))

ax=sns.barplot(data=casos_urbanos ,x='porcentaje_poblacion_urbana', y=casos_urbanos.index,label='Porcentaje de poblaciÃ³n urbana')

for index, value in enumerate(casos_urbanos['porcentaje_poblacion_urbana']):
    ax.text(value, index, f'{value*100:.2f}%', color='black', ha="left", va="center")

ax2 = sns.barplot(x='porcentaje_poblacion_confirmada', y=casos_urbanos.index, data=casos_urbanos,label='Porcentaje de casos confirmados')

for index, value in enumerate(casos_urbanos['porcentaje_poblacion_confirmada']):
    ax2.text(value, index, f'{value*100:.2f}%', color='orange', ha="left", va="center")
 
    
ax.set_xlabel('Porcentaje')
ax.set_ylabel('PaÃ­s')
ax.set_title('UrbanizaciÃ³n y Confirmados en Latinoamerica')
ax.legend(bbox_to_anchor=(1.32, 1), loc="upper right")
plt.tight_layout()

```

El pais con mayor impacto en la propagaciÃ³n fue Chile,en los anÃ¡lisis anteriores se concluyÃ³ que eES uno de los paÃ­ses con mejor manejo de la pandemia, por lo que con este nuevo anÃ¡lisis se concluye que el alto nÃºmero de contagiados se debe a su alta densidad poblacional.

Para comprender mÃ¡s certeramente cuales son los factores influyentes en la mortalidad debido al COVID, se analizan los datos de los paises con mayor tasa de mortalidad debido al virus, y se agrupa su poblaciÃ³n en distintos grupos etarios.

```python
df_latinoamerica_filtrada['porcentaje_mortalidad']=df_latinoamerica_filtrada['cumulative_deceased']/df_latinoamerica_filtrada['population']
alta_mortalidad_paises= df_latinoamerica_filtrada.groupby('country_name')['porcentaje_mortalidad'].max().sort_values(ascending=False).head(5).index
grupos_etarios = ['population_age_00_09', 'population_age_10_19', 'population_age_20_29', 'population_age_30_39', 'population_age_40_49', 'population_age_50_59', 'population_age_60_69', 'population_age_70_79', 'population_age_80_and_older']
distribucion_edad = df_latinoamerica_filtrada[df_latinoamerica_filtrada['country_name'].isin(alta_mortalidad_paises)].groupby('country_name')[grupos_etarios].mean()
paises = distribucion_edad.index.unique()
```

Para tener un mejor entendimiento de los efectos de la pandemia en los distintos paÃ­ses, se procede a una comparaciÃ³n de las dosis suministradas en promedio por habitantes en cada paÃ­s. 

```python
#Comparacion de estrategias de vacunacion por pais
estrategias_vacunacion = df_latinoamerica_filtrada.groupby('country_name')['cumulative_vaccine_doses_administered'].max() / df_latinoamerica_filtrada.groupby('country_name')['population'].mean()
estrategias_vacunacion.sort_values(ascending=False).plot(kind='bar', figsize=(12, 8), color='purple', grid=True)

plt.xlabel('PaÃ­s')
plt.ylabel('Dosis Administradas por Habitante')
plt.xticks(rotation=45)
plt.title('ComparaciÃ³n de Estrategias de VacunaciÃ³n en AmÃ©rica Latina')
```

En consonancia con los anÃ¡lisis anteriores se concluye que el paÃ­s con mÃ¡s dÃ³sis suministradas es Chile, lo que explica su mejor manejo de la pandemia. Seguido de PerÃº y Argentina. Del grÃ¡fico tambiÃ©n se concluye que Brasil es el paÃ­s con menos dÃ³sis suministradas por habitante, lo que podrÃ­a explicar su alta tasa de mortalidad. Por lo que hay una clara tendencia en que Brasil, y Colombia son los paÃ­ses que necesitan mÃ¡s ayuda  y en donde la creaciÃ³n de nuevos centros serÃ­a un Ã©xito.

TambiÃ©n es Ãºtil conocer como afectan enfermedades preexistentes como la diabetes o el habito de fumar, en los paÃ­ses con altas tasas de mortalidad y en los paÃ­ses con baja tasa de mortalidad.

```python
#Prevalencia de condiciones preexistentes en PaÃ­ses con altas y bajas tasas de mortalidad
#Se seleccionan los 3 paises con mayor tasa de mortalidad
paises_mayor_mortalidad = df_latinoamerica_filtrada.groupby('country_name')['porcentaje_mortalidad'].max().nlargest(3).index
#Se buscan los 3 paises con menos muertes
paises_menos_muertes = df_latinoamerica_filtrada.groupby('country_name')['porcentaje_mortalidad'].max().nsmallest(3).index
#Organizacion de mortalidad de los paises
mortalidad_paises=df_latinoamerica_filtrada.groupby('country_name')['porcentaje_mortalidad'].max().sort_values(ascending=False)


```

```python
datos_superiores = df_latinoamerica_filtrada[df_latinoamerica_filtrada['country_name'].isin(paises_mayor_mortalidad)]
datos_inferiores = df_latinoamerica_filtrada[df_latinoamerica_filtrada['country_name'].isin(paises_menos_muertes)]

fig,ax = plt.subplots(
    ncols=2,
    nrows=2,
    figsize=(14, 16)
)

# Prevalencia de diabetes en top
sns.barplot(data=datos_superiores, x='country_name', y='diabetes_prevalence', hue='country_name', ax=ax[0][0])
ax[0][0].set_xlabel('PaÃ­s')
ax[0][0].set_ylabel('Prevalencia de Diabetes (%)')
ax[0][0].set_title('Prevalencia de Diabetes en PaÃ­ses con Altas Tasas de Mortalidad por COVID-19')

# Prevalencia de fumar en top
sns.barplot(data=datos_superiores, x='country_name', y='smoking_prevalence', hue='country_name', ax=ax[0][1])
ax[0][1].set_xlabel('PaÃ­s')
ax[0][1].set_ylabel('Prevalencia de Fumar (%)')
ax[0][1].set_title('Prevalencia de Fumar en PaÃ­ses con Altas Tasas de Mortalidad por COVID-19')

# Prevalencia de diabetes en top
sns.barplot(data=datos_inferiores, x='country_name', y='diabetes_prevalence', hue='country_name', ax=ax[1][0])
ax[1][0].set_xlabel('PaÃ­s')
ax[1][0].set_ylabel('Prevalencia de Diabetes (%)')
ax[1][0].set_title('Prevalencia de Diabetes en PaÃ­ses con Bajas Tasas de Mortalidad por COVID-19')

# Prevalencia de fumar en top
sns.barplot(data=datos_inferiores, x='country_name', y='smoking_prevalence', hue='country_name', ax=ax[1][1])
ax[1][1].set_xlabel('PaÃ­s')
ax[1][1].set_ylabel('Prevalencia de Fumar (%)')
ax[1][1].set_title('Prevalencia de Fumar en PaÃ­ses con Bajas Tasas de Mortalidad por COVID-19')

plt.tight_layout()
```

A partir del grÃ¡fico se identifica que la diabetes afecta mucho mas a MÃ©xico y Brasil; y el fumar afecta mÃ¡s a Chile y Argentina.

Sin embargo mientras Brasil aparece entre los paÃ­ses con mayor tasa de mortalidad, MÃ©xico aparece entre los mÃ©nores. 
Lo mismo pasas entre Chile (mayor) y Argentina (menor)

Por lo que dificilmente se pueda entender una correlaciÃ³n entre las situaciones preexistentes y la mortalidad por COVID. 

Conociendo que la mortalidad no es afectada por factores preexistentes,conocer si la mortalidad es igual en todos los paÃ­ses es importante para realizar anÃ¡lisis de mayor profundidad.

```python
#Fatalidad por pais
df_latinoamerica_filtrada['ratio_fatalidad'] = df_latinoamerica_filtrada['cumulative_deceased'] / df_latinoamerica_filtrada['cumulative_confirmed'] 
ratio_fatalidad = df_latinoamerica_filtrada.groupby('country_name')['ratio_fatalidad'].last().sort_values()

plt.figure(figsize=(12, 8))
ratio_fatalidad.plot(kind='bar', color='purple')
plt.xlabel('PaÃ­s')
plt.ylabel('Tasa de Letalidad')
plt.title('ComparaciÃ³n de la tasa de letalidad del COVID-19 entre PaÃ­ses de AmÃ©rica Latina')
```

Se concluye que la enfermedad ha tenido el mismo comportamiento de fatalidad en los distintos paÃ­ses. 
Para terminar el anÃ¡lisis se busca conocer como ha variado la letalidad de la enfermedad en el tiempo. 

```python
#Evolucion de muertes 
plt.figure(figsize=(14, 7))
sns.lineplot(
    data=df_latinoamerica_filtrada, 
    x='date', 
    y='ratio_fatalidad', 
    estimator='mean'
)
plt.title('AnÃ¡lisis Temporal de la Mortalidad por COVID-19 en AmÃ©rica Latina')
plt.xlabel('Fecha')
plt.ylabel('Tasa de Mortalidad')
```

Se concluye que la tasa de mortalidad de la enfermedad ha disminuido posiblemente, debido a la implementaciÃ³n de la vacunaciÃ³n en los distintos paÃ­ses
